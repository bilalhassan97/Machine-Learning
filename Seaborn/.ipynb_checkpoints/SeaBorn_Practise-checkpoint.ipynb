```python
a=sns.load_dataset('flights')
sns.relplot(x='passengers',y='month',data=a)
```

```python
sns.relplot(x='passengers',y='month',hue='year',data=a)
```

```python
b=sns.load_dataset('tips')
sns.relplot(x='tip',y='total_bill',data=b,kind='line')
```

```python
sns.relplot(x='day',y='total_bill',data=b,kind='line')
```

```python
sns.catplot(x="day",y="total_bill",data=b)
```

```python
sns.violinplot(x='day',y='total_bill',hue='sex',data=b,split='true')
```

```python
sns.boxplot(x='day',y='total_bill',hue="sex",data=b)
```

```python
from scipy import stats

c=np.random.normal(loc=5,size=100,scale=2)

sns.distplot(c)
```
